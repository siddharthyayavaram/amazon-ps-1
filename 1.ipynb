```python
def summary(df):
    print(f"Dataset Shape: {df.shape}")
    summary= pd.DataFrame(df.dtypes, columns=['dtypes'])
    summary= summary.reset_index()
    summary['Feature Name'] = summary['index']
    summary = summary[['Feature Name', 'dtypes']]
    summary['missing'] = df.isnull().sum().values
    summary['Uniques'] = df.nunique().values
    return summary

summary(df)
```

```python
def remove_outliers_iqr(dataframe, column):
    data = dataframe[column]
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    cleaned_dataframe = dataframe[(data > lower_bound) & (data < upper_bound)]
    return cleaned_dataframe
```

```python
df_new = remove_outliers_iqr(df_new, 'Height')
df_new = remove_outliers_iqr(df_new, 'Width')
df_new = remove_outliers_iqr(df_new, 'Weight')
df_new = remove_outliers_iqr(df_new, 'Price Of Sculpture')
```

```python
X = df_new.drop('Cost', axis = 1)
y = df_new['Cost']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state = 42)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
```

```python
xgb = XGBRegressor()
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 6],
    'n_estimators': [100, 200],
}

xgb_grid = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5)
xgb_grid.fit(X_train, y_train)

y_pred = xgb_grid.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

n = len(y_test)  # Number of samples
p = X_test.shape[1]  # Number of predictors

adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

accuracy_score = adjusted_r2 * 100
print('Accuracy Score:', accuracy_score)


print('Best parameters:', xgb_grid.best_params_)
print('RMSE:', rmse)
print('R-squared:', r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

rf = RandomForestRegressor()
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 3, 6]
}

rf_grid = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=5)
rf_grid.fit(X_train, y_train)

y_pred = rf_grid.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

adjusted_r2 = 1 - (1 - r2) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
accuracy_score = adjusted_r2 * 100

print('Accuracy Score:', accuracy_score)

print('Best parameters:', rf_grid.best_params_)
print('RMSE:', rmse)
print('R-squared:', r2)
```
